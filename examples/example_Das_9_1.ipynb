```python
import pandas as pd

from geotechpy.soil_profile import SoilProfile

# Show all columns
pd.set_option("display.max_columns", None)
```

```python
# Read the soil profile data from a CSV file
sp_df = pd.read_csv("soil_profile_Das_9_1.csv")

sp_df
```

```python
# Use the SoilProfile class to calculate the total stress, pore water pressure, and effective stress
# The SoilProfile class requires the dataframe with the soil profile data, surcharge load, water surface elevation, and the units system being used.

# Declare the surcharge load in kPa or psf
surcharge_load = 0

# Declare the water surface elevation in m or ft
water_surface_elevation = 94

# Declare the units system as either "metric" or "imperial"
units = "metric"

# Create an instance of the SoilProfile class
soil_profile = SoilProfile(sp_df, surcharge_load, water_surface_elevation, units)

# To calculate the total stress, pore water pressure, and effective stress, use the calculate_all() method
soil_profile.calculate_all()

# Display the results
soil_profile.dataframe
```

```python
# To plot the vertical stress profile, use the plot_profile() method
soil_profile.plot_profile()
```
